# Phase 1 EDA: Massive (Polygon) SPY Options

Quick coverage/missingness plots for normalized data produced in Phase 1.


```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

pd.set_option("display.max_rows", 20)
DATA_PATH = Path("../data/processed/options.parquet")
assert DATA_PATH.exists(), "Run pipelines.phase1.normalize first to create processed data"

df = pd.read_parquet(DATA_PATH)
df.head()
```

```python
coverage = df.groupby(["as_of", "expiration"]).size().rename("rows").reset_index()
coverage_pivot = coverage.pivot(index="expiration", columns="as_of", values="rows")
coverage_pivot.fillna(0).astype(int)

```

```python
fig, ax = plt.subplots(figsize=(10, 6))
df.groupby("as_of")[["bid", "ask", "mid"]].count().plot(kind="bar", ax=ax)
ax.set_title("Quote field counts per as_of date")
ax.set_ylabel("Count")
plt.tight_layout()
plt.show()

```
